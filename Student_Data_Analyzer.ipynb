```
import numpy as np

import matplotlib.pyplot as plt
```

```
# Create a 2D NumPy array called 'marks'
# Each row represents one student
# Each column represents one subject
marks = np.array([
    [85, 78, 92],   # Student 1 marks
    [70, 65, 80],   # Student 2 marks
    [90, 88, 91],   # Student 3 marks
    [60, 30, 68],   # Student 4 marks
    [75, 93, 37]    # Student 5 marks
])


# Print the shape (rows and columns) of the array
# Output will be (5, 3)
# 5 rows = 5 students
# 3 columns = 3 subjects
print(marks.shape)


# Print the complete array
# This shows all student marks
print(marks)


# Create a list containing student names
# These names will appear on the X-axis
students = ["S1", "S2", "S3", "S4", "S5"]


# Plot line chart for Subject 1
# marks[:, 0] means:
# :  -> select all rows
# 0  -> select column index 0 (first subject)
# marker='o' adds circle points on the line
# label is used in the legend
plt.plot(students, marks[:, 0], marker='o', label="Subject 1")


# Plot line chart for Subject 2
# marks[:, 1] selects second column
plt.plot(students, marks[:, 1], marker='o', label="Subject 2")


# Plot line chart for Subject 3
# marks[:, 2] selects third column
plt.plot(students, marks[:, 2], marker='o', label="Subject 3")


# Set label for X-axis
# X-axis contains student names
plt.xlabel("Students")


# Set label for Y-axis
# Y-axis contains marks
plt.ylabel("Marks")


# Add title to the chart
plt.title("Student Marks Line Chart")


# Show legend box
# Legend explains which line belongs to which subject
plt.legend()


# Display the final chart window
plt.show()
```

```
# Calculate average marks for each student (row-wise)
Student_avg = np.mean(marks, axis=1)
# np.mean() → finds average
# marks → your data (2D array: rows = students, columns = subjects)
# axis=1 → go across rows (-->), meaning each student separately

# Print the result
print("Avg Per Student", Student_avg)
# Displays the text + the calculated averages



students = ["s1","s2","s3","s4","s5"]

# Bar chart
plt.bar(students, Student_avg)

# Labels and title
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.title("Average Marks Per Student")

# Show chart
plt.show()
```

```
# Calculate average marks for each subject (column-wise)
subject_avg = np.mean(marks, axis=0)
# np.mean() → finds average
# marks → 2D array (rows = students, columns = subjects)
# axis=0 → go down columns (⬇), meaning each subject separately

# Print the result
print("Subject Avg : ", subject_avg)
# Displays subject-wise averages

# Subject names
subjects = ["Subject 1", "Subject 2", "Subject 3"]


students = ["S1","S2","S3","S4","S5"]

plt.bar(subjects,subject_avg);

# Create bar chart
plt.bar(subjects, subject_avg)

# Labels and title
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.title("Average Marks Per Subject")

# Show chart
plt.show()
```

```
# find highest and lowest value from the 2d array

highest_marks = np.max(marks)
lowest_marks = np.min(marks)

print("higherst mark :",highest_marks)
print("lowest mark : ",lowest_marks)

# Create labels for pie chart
labels = ["Highest Mark", "Lowest Mark"]

# Create values list for pie chart
values = [highest_marks, lowest_marks]


# Create pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%')


# Add chart title
plt.title("Highest vs Lowest Marks")


# Display chart
plt.show()
```

```
# Find highest mark for each student (row-wise)
best_per_student = np.max(marks, axis=1)
# np.max() → finds maximum value
# axis=1 → go across rows (each student)

# Print result
print("Best mark per student:", best_per_student)

# Student names
students = ["S1", "S2", "S3", "S4", "S5"]

# Subject names
subjects = ["Subject 1", "Subject 2", "Subject 3"]

# return best subjuct index value
best_subject_index = np.argmax(marks, axis=1)

# Create horizontal bar chart
plt.barh(students, best_per_student)


for i in range(len(students)):

    # Get subject name
    subject_name = subjects[best_subject_index[i]]

    # Put text INSIDE the bar
    plt.text(
        best_per_student[i] / 2,   # Middle of bar
        i,                         # Bar position
        subject_name,              # Text
        ha='center',               # Horizontal align
        va='center',               # Vertical align
        color='white'              # Text color
    )

# Labels and title
plt.xlabel("Highest Marks")
plt.ylabel("Students")
plt.title("Best Mark Per Student")

# Show chart
plt.show()
```

```
# =========================================================
# CHECK PASS / FAIL STATUS
# =========================================================

# Check pass/fail status for each student
pass_status = np.all(marks >= 40, axis=1)

# marks >= 40
# Checks every mark
# Returns True if mark >= 40
# Returns False if mark < 40

# np.all(..., axis=1)
# Checks row-wise
# Student passes only if ALL subjects are passed

# Example:
# [85, 78, 92] -> True
# [75, 93, 37] -> False (because 37 < 40)


# Print pass/fail result
print("Pass status:", pass_status)


# =========================================================
# HEATMAP CHART
# =========================================================

# Convert Boolean values into numbers
# True  -> 1
# False -> 0
status_numeric = pass_status.astype(int)


# Create student labels
students = ["S1", "S2", "S3", "S4", "S5"]


# Create figure size
plt.figure(figsize=(8, 2))


# Create heatmap
# [status_numeric] converts 1D array into 2D shape
plt.imshow([status_numeric], aspect='auto')


# Add student labels on X-axis
plt.xticks(range(len(students)), students)


# Remove Y-axis ticks
plt.yticks([])


# Add text inside heatmap cells
for i in range(len(students)):

    # If student passed
    if pass_status[i]:
        text = "Pass"

    # If student failed
    else:
        text = "Fail"

    # Add text inside each box
    plt.text(
        i,                  # X position
        0,                  # Y position
        text,               # Text to display
        ha='center',        # Horizontal alignment
        va='center',        # Vertical alignment
        color='black'       # Text color
    )


# Add chart title
plt.title("Student Pass/Fail Heatmap")


# Display heatmap chart
plt.show()



# =========================================================
# WAFFLE CHART
# =========================================================

# Count total passed students
pass_count = np.sum(pass_status)

# Count total failed students
fail_count = len(pass_status) - pass_count


# Print counts
print("Passed Students :", pass_count)
print("Failed Students :", fail_count)


# Create new figure
plt.figure(figsize=(8, 2))


# Total number of students
total = len(pass_status)


# Create waffle chart using square markers
for i in range(total):

    # Passed students
    if i < pass_count:

        plt.scatter(
            i,              # X position
            0,              # Y position
            s=1500,         # Square size
            marker='s',     # Square shape
            color='green'
        )

        # Add text inside square
        plt.text(
            i,
            0,
            "Pass",
            ha='center',
            va='center',
            color='white'
        )

    # Failed students
    else:

        plt.scatter(
            i,
            0,
            s=1500,
            marker='s',
            color='red'
        )

        # Add text inside square
        plt.text(
            i,
            0,
            "Fail",
            ha='center',
            va='center',
            color='white'
        )


# Remove axis lines and numbers
plt.axis('off')


# Add chart title
plt.title("Student Pass/Fail Waffle Chart")


# Show waffle chart
plt.show()
```

```
#find topper
total_marks = np.sum(marks, axis=1)
topper_index = np.argmax(total_marks)
print("Topper is student index:", topper_index)

students = ["S1", "S2", "S3", "S4", "S5"]

# =========================================================
# CREATE BAR COLORS
# =========================================================

# Default color for all students
colors = ["skyblue", "skyblue", "skyblue", "skyblue", "skyblue"]

# Change topper bar color
colors[topper_index] = "gold"

# =========================================================
# CREATE HIGHLIGHTED BAR CHART
# =========================================================

# Create figure size
plt.figure(figsize=(8, 5))


# Create bar chart
bars = plt.bar(students, total_marks, color=colors)


# =========================================================
# ADD TOTAL MARKS ON TOP OF BARS
# =========================================================

for i in range(len(students)):

    plt.text(
        i,                          # X position
        total_marks[i] + 2,         # Y position
        total_marks[i],             # Text value
        ha='center'                 # Horizontal alignment
    )


# =========================================================
# ADD TOPPER LABEL
# =========================================================

# Add TOPPER text inside the topper bar
plt.text(
    topper_index,                         # X position (topper bar)
    total_marks[topper_index] / 2,       # Middle height of bar
    "TOPPER",                             # Text to display
    ha='center',                          # Horizontal alignment
    va='center',                          # Vertical alignment
    color='white',                        # Text color
    fontsize=12                           # Text size
)


# =========================================================
# ADD LABELS AND TITLE
# =========================================================

plt.xlabel("Students")
plt.ylabel("Total Marks")
plt.title("Topper Student Highlighted Bar Chart")


# =========================================================
# SHOW CHART
# =========================================================

plt.show()
```

```
# =========================================================
# ASSIGN GRADES
# =========================================================

grades = np.where(
    Student_avg >= 85, "A",

    np.where(
        Student_avg >= 70, "B", "C"
    )
)

# If average >= 85  -> Grade A
# Else if average >= 70 -> Grade B
# Else -> Grade C


# =========================================================
# PRINT RESULTS
# =========================================================

print("Student Average:", Student_avg)
print("Grades:", grades)


# =========================================================
# CREATE STUDENT LABELS
# =========================================================

students = ["S1", "S2", "S3", "S4", "S5"]


# =========================================================
# ASSIGN COLORS BASED ON GRADES
# =========================================================

colors = []


# Loop through each grade
for grade in grades:

    # Grade A students
    if grade == "A":
        colors.append("green")

    # Grade B students
    elif grade == "B":
        colors.append("blue")

    # Grade C students
    else:
        colors.append("red")


# =========================================================
# CREATE COLOR-CODED BAR CHART
# =========================================================

# Create chart size
plt.figure(figsize=(8, 5))


# Create bar chart
plt.bar(
    students,          # X-axis labels
    Student_avg,       # Bar heights
    color=colors       # Bar colors
)


# =========================================================
# ADD GRADE TEXT INSIDE BARS
# =========================================================

for i in range(len(students)):

    plt.text(
        i,                         # X position
        Student_avg[i] / 2,       # Middle height of bar
        grades[i],                # Grade text
        ha='center',              # Horizontal alignment
        va='center',              # Vertical alignment
        color='white',            # Text color
        fontsize=14               # Text size
    )


# =========================================================
# ADD LABELS AND TITLE
# =========================================================

plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.title("Student Grades Color-Coded Bar Chart")


# =========================================================
# SHOW BAR CHART
# =========================================================

plt.show()



# =========================================================
# COUNT TOTAL GRADES
# =========================================================

# Count Grade A students
A_count = np.sum(grades == "A")

# Count Grade B students
B_count = np.sum(grades == "B")

# Count Grade C students
C_count = np.sum(grades == "C")


# Print grade counts
print("Grade A:", A_count)
print("Grade B:", B_count)
print("Grade C:", C_count)


# =========================================================
# CREATE LABELS AND VALUES FOR DONUT CHART
# =========================================================

labels = ["Grade A", "Grade B", "Grade C"]

values = [A_count, B_count, C_count]


# =========================================================
# CREATE DONUT CHART
# =========================================================

# Create figure size
plt.figure(figsize=(6, 6))


# Create pie chart
plt.pie(
    values,
    labels=labels,
    autopct='%1.1f%%'
)


# =========================================================
# CREATE WHITE CIRCLE IN CENTER
# =========================================================

centre_circle = plt.Circle(
    (0, 0),       # Center position
    0.30,         # Circle radius
    fc='white'    # Fill color
)


# Add circle to chart
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


# =========================================================
# ADD TITLE
# =========================================================

plt.title("Student Grade Distribution Donut Chart")


# =========================================================
# SHOW DONUT CHART
# =========================================================

plt.show()
```

```


# =========================================================
# COUNT PASSED STUDENTS
# =========================================================

# Count total passed students
num_passed = np.sum(pass_status)

# Print result
print("Number of students passed:", num_passed)


# =========================================================
# CALCULATE PASS PERCENTAGE
# =========================================================

# Total students
total_students = len(pass_status)

# Calculate percentage
pass_percentage = (num_passed / total_students) * 100

# Print percentage
print("Pass Percentage:", pass_percentage)


# =========================================================
# CREATE GAUGE CHART
# =========================================================

# Create figure size
plt.figure(figsize=(8, 4))


# Gauge values
values = [
    pass_percentage,
    100 - pass_percentage
]


# Gauge colors
colors = ["green", "lightgray"]


# Create donut-style pie chart
plt.pie(
    values,
    startangle=180,
    counterclock=False,
    colors=colors,
    wedgeprops={
        'width': 0.4
    }
)


# =========================================================
# HIDE LOWER HALF
# =========================================================

# Cover bottom part to create gauge look
plt.fill_between(
    [-2, 2],
    -2,
    0,
    color='white'
)


# =========================================================
# ADD CENTER TEXT
# =========================================================

plt.text(
    0,
    0.1,
    f"{pass_percentage:.1f}%",
    ha='center',
    fontsize=20
)

plt.text(
    0,
    -0.1,
    "Students Passed",
    ha='center',
    fontsize=12
)


# =========================================================
# ADD TITLE
# =========================================================

plt.title("Student Pass Percentage Gauge Chart")


# =========================================================
# SHOW CHART
# =========================================================

plt.show()
```

```
# count  fail
num_failed = len(pass_status) - num_passed
print("Number of students failed:", num_failed)

# Total students
total_students = len(pass_status)

# Calculate percentage
fail_percentage = (num_failed / total_students) * 100

# Print percentage
print("fail Percentage:", fail_percentage)

# =========================================================
# CREATE GAUGE CHART
# =========================================================

# Create figure size
plt.figure(figsize=(8, 4))


# Gauge values
values = [
    fail_percentage,
    100 - pass_percentage
]


# Gauge colors
colors = ["green", "lightgray"]


# Create donut-style pie chart
plt.pie(
    values,
    startangle=180,
    counterclock=False,
    colors=colors,
    wedgeprops={
        'width': 0.4
    }
)


# =========================================================
# HIDE LOWER HALF
# =========================================================

# Cover bottom part to create gauge look
plt.fill_between(
    [-2, 2],
    -2,
    0,
    color='white'
)


# =========================================================
# ADD CENTER TEXT
# =========================================================

plt.text(
    0,
    0.1,
    f"{fail_percentage:.1f}%",
    ha='center',
    fontsize=20
)

plt.text(
    0,
    -0.1,
    "Students Failed",
    ha='center',
    fontsize=12
)


# =========================================================
# ADD TITLE
# =========================================================

plt.title("Student Fail Percentage Gauge Chart")


# =========================================================
# SHOW CHART
# =========================================================

plt.show()


```
